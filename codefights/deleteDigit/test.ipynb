```python
def deleteDigit(n):    
    st = str(n)
    values = []
    if len(st) == 2:
        return int(max(list(st)))
    for i in range(len(st)):
        value = st[0:i] + st[i+1:]
        values.append(value)
        # print(values)
    return int(max(values))
```
